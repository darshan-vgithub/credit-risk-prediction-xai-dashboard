```python
import pandas as pd

columns = [
    "CheckingAccount",
    "Duration",
    "CreditHistory",
    "Purpose",
    "CreditAmount",
    "SavingsAccount",
    "EmploymentSince",
    "InstallmentRate",
    "PersonalStatusSex",
    "OtherDebtors",
    "ResidenceSince",
    "Property",
    "Age",
    "OtherInstallmentPlans",
    "Housing",
    "ExistingCredits",
    "Job",
    "Dependents",
    "Telephone",
    "ForeignWorker",
    "Risk"
]

df = pd.read_csv(
    "../data/raw/german.data",
    sep=r"\s+",
    header=None,
    names=columns
)
```

```python
df.iloc[0]
```

Data Dictionary

| Feature         | Description                       |
| --------------- | --------------------------------- |
| CheckingAccount | Status of checking account        |
| Duration        | Loan duration in months           |
| CreditHistory   | Previous credit repayment history |
| Purpose         | Reason for loan                   |
| CreditAmount    | Loan amount requested             |
| SavingsAccount  | Savings level                     |
| EmploymentSince | Employment duration               |
| InstallmentRate | Installment percentage of income  |
| Age             | Customer age                      |
| Housing         | Housing status                    |
| ExistingCredits | Existing loans                    |
| Job             | Employment category               |
| Risk            | Good or Bad credit                |
