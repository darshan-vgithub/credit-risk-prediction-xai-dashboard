```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
columns = [
    "CheckingAccount",
    "Duration",
    "CreditHistory",
    "Purpose",
    "CreditAmount",
    "SavingsAccount",
    "EmploymentSince",
    "InstallmentRate",
    "PersonalStatusSex",
    "OtherDebtors",
    "ResidenceSince",
    "Property",
    "Age",
    "OtherInstallmentPlans",
    "Housing",
    "ExistingCredits",
    "Job",
    "Dependents",
    "Telephone",
    "ForeignWorker",
    "Risk"
]
```

```python
df = pd.read_csv(
    "../data/raw/german.data",
    sep=r"\s+",
    header=None,
    names=columns
)
```

```python
# Numerical Feature Analysis
```

```python
numerical_cols = [
    "Duration",
    "CreditAmount",
    "InstallmentRate",
    "ResidenceSince",
    "Age",
    "ExistingCredits",
    "Dependents"
]

df[numerical_cols].describe()
```

```python
df[numerical_cols].hist(
    figsize=(15,10),
    bins=20
)

plt.tight_layout()
plt.show()
```

```python
df[numerical_cols].skew()
```

```python
plt.figure(figsize=(15,8))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```
