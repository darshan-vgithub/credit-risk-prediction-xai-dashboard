```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(
    "../data/raw/german.data",
    sep=r"\s+",
    header=None,
    names=columns
)
```

```python
# Numerical Feature Analysis
```

```python
numerical_cols = [
    "Duration",
    "CreditAmount",
    "InstallmentRate",
    "ResidenceSince",
    "Age",
    "ExistingCredits",
    "Dependents"
]

df[numerical_cols].describe()
```

```python
df[numerical_cols].hist(
    figsize=(15,10),
    bins=20
)

plt.tight_layout()
plt.show()
```

```python
df[numerical_cols].skew()
```

```python
plt.figure(figsize=(15,8))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
categorical_cols = [
    "CheckingAccount",
    "CreditHistory",
    "Purpose",
    "SavingsAccount",
    "EmploymentSince",
    "PersonalStatusSex",
    "OtherDebtors",
    "Property",
    "OtherInstallmentPlans",
    "Housing",
    "Job",
    "Telephone",
    "ForeignWorker"
]
```

```python
for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].value_counts())
```

```python
for col in categorical_cols:

    plt.figure(figsize=(6,4))

    sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index
    )

    plt.title(col)
    plt.xticks(rotation=45)

    plt.show()
```
